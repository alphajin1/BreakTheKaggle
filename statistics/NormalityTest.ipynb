```python
from scipy import stats
import seaborn as sns

r = stats.norm.rvs(size=1000)
sns.distplot(r)
```

```python
from scipy.stats import shapiro

# 정규성 검정 01 (샤피로 검정)
# 귀무가설 : 표본의 모집단이 정규분포를 따른다.
# 0.05보다 낮으니, 정규성 검정을 통과했다.
shapiro(r)
```

```python
from scipy.stats import kstest

# 정규성 검정 02 (콜모고로프 스미르노프 검정)
# 귀무가설 : 두 집단은 같다. 
# --> 기각 실패, 두 집단은 같다고 볼 수 있다. 
kstest(r, "norm")
```

```python
from scipy.stats import probplot
import matplotlib.pyplot as plt 
import numpy as np 

# 정규성 검정 03 (QQ-Plot)
# 직선에 가깝다면, 정규성을 따른다고 볼 수 있다.
plt.figure(figsize=(7,7))
stats.probplot(r, plot=plt)
plt.axis("equal")
plt.show()
```
