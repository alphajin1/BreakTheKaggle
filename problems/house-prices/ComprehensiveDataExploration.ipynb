```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df_train = pd.read_csv('data/train.csv')
df_train.columns
```

## EDA - SalePrice
- 우리가 예측해야할 값

```python
sns.distplot(df_train['SalePrice'])
```

```python
# 정규분포가 아니다!
# appreciable positive skewness(왜도) 가 있음
# Kurtosis(첨도) 가 있음
print("Skewness: %f" % df_train['SalePrice'].skew())
print("Kurtosis: %f" % df_train['SalePrice'].kurt())
```

### Relationship with numerical variables

```python
#scatter plot grlivarea/saleprice
var = 'GrLivArea' # 지상위의 평수
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
```

```python
#scatter plot totalbsmtsf/saleprice
var = 'TotalBsmtSF' # 지하 면적의 평수
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
# exponential relation (?)
# 때때로 TotalBsmtSF가 0원짜리 SalePrice와 관련이 있다?
```

### Relationship with categorical features

```python
#box plot overallqual/saleprice
var = 'OverallQual' # 집의 전체 재료 및 마무리 등급 / Rates the overall material and finish of the house 
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
```

```python
var = 'YearBuilt' # 건축년도
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(16, 8))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
plt.xticks(rotation=90);
```
