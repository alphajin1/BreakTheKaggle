## moment 란? 간략하게는 분포를 이해하는데 도와주는 값이다.
- Average : The first moment of the data or population.
- Varience : The second moment of the data or population.
- Skewness : The third moment of the data or population.
- Kurtosis : The forth moment of the data or population.

```python
import pandas as pd
df = pd.read_csv('data/train.csv')
pd.DataFrame(df.skew()).T
```

## Skewness (s)
- s < -1 or s > 1 이면, 높은 Skewness
- -1 ~ s < 0.5 or 0.5 < s < 1 적당한 Skewness
- -0.5 ~ s < 0.5 거이 Symmetric

```python
import seaborn as sns
print(df['LotArea'].skew())
sns.distplot(df['LotArea'])
```

```python
import seaborn as sns
print(df['YearBuilt'].skew())
sns.distplot(df['YearBuilt'])
```

## Kurtosis (첨도) (주의:peakedness에 대한 measure가 아니다. )
- Outlier에 대한 정보를 나타낸다. (얼마나, rare, extreme value가 존재하는지)
- Fisher's definition of kurtosis 에 따라, 0이면 Normal Distribution
- 0보다 크면 Normal Distribution보다 Outlier가 많다.
- Kurtosis값이 0보다 작다는 것은 정규분포보다 Outlier부분이 적다는 것을 의미한다.

```python
import seaborn as sns
print(df['LotArea'].kurt())
sns.distplot(df['LotArea'])
```

```python
import seaborn as sns
print(df['YearRemodAdd'].kurt())
sns.distplot(df['YearRemodAdd'])
```

## 정규분포(Normal Distribution)의 경우 Skewness, Kurtosis의 값은 0에 가깝지만, 0에 가까운 Skewness, Kurtois값을 가졌다고 해서 Normal Distribution이라고 할 수 있을까?
- 아니다. 정규분포에 맞는지 점검을 해야한다. 
- Pearson chi square test ... 등 다른 정규성 검증 방법을 이용하자.

```python
# Normal distribution
import numpy as np
from math import sqrt
n = np.random.normal(100 * 0.5, sqrt(100 * 0.5*0.5), size=10000)
print(pd.DataFrame(n).kurt())
print(pd.DataFrame(n).skew())

sns.distplot(pd.DataFrame(n))
```

## SalePrice 의 Skewness, Kurtosis 는? 

```python
import pandas as pd

df = pd.read_csv('data/train.csv')
print("Skewness: %f" % df['SalePrice'].skew())
print("Kurtosis: %f" % df['SalePrice'].kurt())
```

```python
import seaborn as sns
sns.distplot(df['SalePrice'])
```
